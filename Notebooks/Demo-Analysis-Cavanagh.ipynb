In this notebook, we show how to use the functions in ../src/utils.py to retrieve the wavelet estimate of DDM, and produce the plots presented in the reference paper, for the Cavanagh dataset.

### 0. Import the functions

```python
import sys
sys.path.insert(1, '../src')

from utils import *
```

### 1. Import the data

In this section, we import the .npz files for the observables P and DT.

```python
P = np.load("../Data/P_observed_cavanagh.npz")
DT = np.load("../Data/DT_observed_cavanagh.npz")
```

### 2. Retrieve $\lambda$ and u_diff using Wavelet estimator

In this section, we use the functions defined in the src file for retrieving the DDM parameters using wavelet estimator. 

```python
l = get_l(P)
```

```python
l_til = l_tilde(l)[0]
l_til2 = l_tilde(l_til)[0]

#check consistence
l_diff = np.abs(l_til2 - l_til)
np.sum(l_diff)
```

```python
DT_til = DT_tilde(l_til, DT)
```

```python
位, udiff = wavelet(l_til, DT_til)

print("位 : ", 位)
print("udiff : \n", udiff)
```

### 3. Visualization

In this section, we evaluate graphically how the DDM estimated through the wavelet estimator fits the data, by plotting this in contrast with the theoretical DDM. 

```python
plot_P(P, 位, udiff)
```

```python
plot_DT(DT, 位, udiff)
```
